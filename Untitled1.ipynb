```python
# Simple Research Data Analysis
data = [12, 15, 12, 18, 20, 15, 12, 22, 15, 12]

# Calculate Mean (Average)
mean = sum(data) / len(data)

# Count how many times each number appears
frequency = {x: data.count(x) for x in set(data)}

print(f"Research Results:")
print(f"Average Value: {mean}")
print(f"Data Frequency: {frequency}")
```
